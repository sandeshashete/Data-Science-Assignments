# Model Building

```python
import statsmodels.formula.api as smf
model=smf.ols("Delivery_Time~Sorting_Time",data=data).fit()
```

```python
print(model.params)
```

```python
model.tvalues
```

```python
model.pvalues
```

```python
# R squared values
model.rsquared,model.rsquared_adj
```

```python
model.summary()
```
